```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Import Data From Excel to Pandas DataFrame (Extracted From Basketball Reference)

```python
rebound_df = pd.read_csv("//Users//yashwant//Downloads//Rebounding.csv")
```

### Plotting Offense Rebounding % against Defensive Rebounding % for Players Using Seaborn and Plotly

```python
# Setting Font
csfont = {'fontname':'Comic Sans MS'}
hfont = {'fontname':'Helvetica'}

# Setting Theme
sns.set(rc={'axes.facecolor':'lightcyan', 'figure.facecolor':'lavender', 'figure.figsize':(15,10)})

# Creating Scatterplot
plot = sns.scatterplot(data = rebound_df, x = "DRB%", y = "ORB%")

# Adding Player Name Labels
for i, name in enumerate(rebound_df["Player"]):
    plt.text(rebound_df["DRB%"][i] - 0.1, rebound_df["ORB%"][i] + 0.2, name.split()[1])

# Boxing points >= 85th percentile OREB% & DREB% 
plt.plot([14.5, 26.2], [4.3, 4.3], '--',  color = "red")
plt.plot([14.5, 14.5], [4.3, 13.5], '--', color = "red")

# Labels/Annotations
plt.xlabel('DRB%', fontsize = 15)
plt.ylabel('ORB%', fontsize = 15)

plot.set_ylim(0, 13.5)
plot.set_xlim(3.5, 26.2)

plt.title("The Best Guard Rebounders, 2022-23", fontsize = 25, **hfont)
plt.text(17.67, 10, "The most well-rounded rebounders:\n >= 85th percentile DRB% & ORB%", backgroundcolor = "bisque",fontsize = 16)


plt.text(21.9, -1, "Data via Basketball-Reference")
plt.text(2.3, -1, "Twitter: @Mathketball1")
```
